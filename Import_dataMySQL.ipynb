```python
#Import the libraries I need to handle data and connect to the MySql database
import pandas as pd
#create_egine from sqlalchime is used to connection between Python and the MySql database.
from sqlalchemy import create_engine
```

```python
#Create a Database Connection
# "mysql+pymysql://" tells SQLAlchemy to use the PyMySQL driver to connect to a MySQL database.
#root is my username.
#admin is my password.
#localhost means the database is hosted on the same machine.
#amazon_db is the name of the database I want to connect to.
engine = create_engine("mysql+pymysql://root:admin@localhost/amazon_db")
```

```python
#Load the products CSV File into a Data Frame
df_products = pd.read_csv(r"D:\Projects\MySQL\AmazonProducts\amazon_products.csv")
```

```python
#.head(10) displays the first 10 rows of the DataFrame
df_products.head(10)
```

```python
#Load the categories CSV File into a Data Frame
df_categories = pd.read_csv(r"D:\Projects\MySQL\AmazonProducts\amazon_categories.csv")
```

```python
df_categories.head(10)
```

```python
#Upload the data to MySql
#Create the tables with 'to_sql()'
df_products.to_sql('products',con=engine,if_exists='replace', index=False)
df_categories.to_sql('categories',con=engine, if_exists ='replace', index=False)

#Be careful with if_exists='replace', as it will delete and recreate the table, removing all existing data.
#Para adicionar registros sem apagar, use 'append'
```
